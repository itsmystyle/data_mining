```python
import matplotlib.pyplot as plt
%matplotlib inline

x1 = [5.7761804000001575, 12.878882680999595, 36.859607025000514, 473.8961617069999]
y1 = [0.35, 0.3, 0.25, 0.2]
x2 = [0.02465160400060995, 0.06242808699971647, 0.130023252000683, 1.152778916999523, 2.1841389659984998, 14.929468891999932, 125.32406900099886]
y2 = [0.35, 0.3, 0.25, 0.2, 0.15, 0.1, 0.05]

plt.plot(y1, x1)
plt.plot(y2, x2)
plt.ylabel('Execution time(sec)')
plt.xlabel('Minimum support')

plt.legend(['Apriori', 'Eclat'], loc='upper right')

plt.savefig('figure_Apriori_vs_Eclat.png')
plt.show()
```

```python
import math

logx1 = [math.log(i) for i in x1]
logx2 = [math.log(i) for i in x2]

plt.plot(y1, logx1)
plt.plot(y2, logx2)
plt.ylabel('Log-scaled Execution time(sec)')
plt.xlabel('Minimum support')

plt.legend(['Apriori', 'Eclat'], loc='upper right')

plt.savefig('figure_log_Apriori_vs_Eclat.png')
plt.show()
```
